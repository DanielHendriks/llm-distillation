## Refrator extraction

```python
import wandb
api = wandb.Api()

run_name = 'zms8ebdk'

def get_run(name):
    return api.run(f"/hendriks_kit/thesis-daniel/runs/{name}")

run = get_run(run_name)
```

```python
def get_mean_runtime_in_hours(run):
    return run.history()._runtime.iloc[-1].item() / 60 / 60

def get_mean_mem_allocation_in_gb(run):
    system_metrics = run.history(stream="events")
    gpu_metrics = system_metrics.filter(like='system.gpu')
    allocated_gpu_gb = gpu_metrics.filter(like='gpu.0.memoryAllocatedBytes') / 1e9
    return allocated_gpu_gb.mean().item()
```

```python
def get_training_data_name(run):
    dir_name = run.config['output_dir']
    if 'few-shot-positive-rationale' in dir_name:
        return 'unrevised' 
    elif 'few-shot-revision' in dir_name:
        return 'revised' 
    else: 
        raise ValueError(f'Training data name could not be inferred from {dir_name}')
    
def get_training_mode_name(run):
    training_mode = run.config.get('training_mode')
    if 'task_prefix' in training_mode:
        return 'mt'
    elif 'counterfactual_prefix' in training_mode:
        return 'cf'
    elif 'both' in training_mode:
        return 'mt+cf'
    else: 
        raise ValueError(f'Training mode could not be inffered from {training_mode}')
    
    
training_data = get_training_data_name(run)
training_mode = get_training_mode_name(run)
```

```python
def mk_run_label(training_mode, training_data):
    return f'{training_mode}:{training_data}'
```

```python
run_label = mk_run_label(training_mode, training_data)
run_label
```

```python
def get_model_size(run):
    from_pretrained = run.config.get('from_pretrained')
    if 'large' in from_pretrained:
        return 'T5-large'
    elif 'base' in from_pretrained:
        return 'T5-base'
    else: 
        raise ValueError(f'Model size could not be inferred from {from_pretrained}')
    
model_size = get_model_size(run)
```

```python
runs = api.runs("hendriks_kit/thesis-daniel")
alias_to_name = {r.name: r.id for r in runs if r.name in run_alias}
```

```python
run_names = [alias_to_name[alias] for alias in run_alias]
```

```python
assert len(run_names) == len(run_alias)
```

```python
from wandb.errors import CommError
runs_data = []

for run_name in run_names[:]:
    try: 
        run = get_run(run_name)
        training_mode = get_training_mode_name(run)
        training_data = get_training_data_name(run)
        runs_data.append({    
            'run_name': run_name,
            'model_size': get_model_size(run),
            'run_label': mk_run_label(training_mode, training_data),
            'mean_runtime': get_mean_runtime_in_hours(run),
            'mean_gpu_allocation': get_mean_mem_allocation_in_gb(run),
        })
    except CommError:
        print(f'No data available for run {run_name}')
        continue
    
    
```

```python
import pandas as pd 
runs_df = pd.DataFrame(runs_data)
```

# Create figure

```python
# only select the four training settings relevant to the paper
relevant_settings = ['mt+cf:unrevised', 'mt+cf:revised', 'cf:unrevised', 'mt:unrevised']
runs_df = runs_df[runs_df.run_label.isin(relevant_settings)]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

plt.style.use("thesis_style.mplstyle")

order = ["cf:unrevised", "mt:unrevised", "mt+cf:unrevised", "mt+cf:revised"]
all_labels = order + [lbl for lbl in runs_df["run_label"].unique() if lbl not in order]

runs_df["run_label"] = pd.Categorical(
    runs_df["run_label"], categories=all_labels, ordered=True
)

runs_df = runs_df.rename(
    columns={'model_size': 'Model size', 'mean_runtime': 'Runtime in Hours', 
             'mean_gpu_allocation': 'GPU Memory Allocation in GB'})

df_melted = runs_df.melt(
    id_vars=["run_label","Model size"],
    value_vars=["Runtime in Hours","GPU Memory Allocation in GB"],
    var_name="metric",
    value_name="value"
)


g = sns.catplot(
    data=df_melted,
    y="run_label",
    x="value",
    col="Model size",
    row="metric",
    kind="bar",
    order=order,  # explicit x-axis order
    sharey=False,
    height=3,
    aspect=2.5/2,
    errorbar=None,
    margin_titles=True,
)

g.set_axis_labels("", "")
g.set_titles("Mean {row_name} | Student: T5-{col_name}")

for ax in g.axes.flat:
    ax.tick_params(axis="x", rotation=0)

for ax in g.axes.flat:
    ax.grid(True, axis='x')  # add x grid

plt.tight_layout()
plt.savefig('secondary_metrics.pdf')
plt.show()
```
